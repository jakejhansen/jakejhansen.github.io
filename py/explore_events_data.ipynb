# Data for D3

```python
cols = ['id', 'source', 'start_time', 'latitude', 'longitude', 'venue_escaped']
for v in venues:
    vname = v.replace('_', ' ')
    selected = df.iloc[df[df.venue_escaped.str.lower().str.contains(vname).astype(bool)].venue_escaped.dropna().index][cols]
    selected.to_csv(os.path.join('for_d3', vname.replace(' ', '_') + '.csv'), index=False)
```

```python
def daily_activity(venue_name):
    return df[df.venue_escaped.str.lower().str.contains(venue_name.lower()).astype(bool)]\
        [['start_time']]\
        .assign(day=lambda df: df.start_time.dt.date)\
        .groupby('day') \
        .count()\
        .sort_index()\
        .rename(columns={'start_time': 'daily_total'})\
        .assign(venue=venue_name)

with open(os.path.join('for_d3', 'daily_event_count.csv'), 'w') as f:
    f.write('day,daily_total,venue\n')
    for v in venues:
        vname = v.replace('_', ' ')
        daily_activity(vname).to_csv(f, header=False)
        
daily_activity('Madison Square Garden').head()
```

```python
def hourly_activity(venue_name):
    return df[df.venue_escaped.str.lower().str.contains(venue_name).astype(bool)]\
        [['start_time']]\
        .assign(hour=lambda df: df.start_time.dt.hour)\
        .groupby('hour') \
        .count()\
        .sort_index()

def hourly_all_venues():
    for v in venues:
        vname = v.replace('_', ' ')
        df_v = hourly_activity(vname)
        counts = []
        print(vname)
        for hour in range(24):
            try:
                counts.append(df_v.loc[hour][0])
            except KeyError:
                counts.append(0)
        with open(os.path.join('for_d3', 'hourly_event_count_' + v + '.csv'), 'w') as f:
            f.write(','.join(map(str, counts)) + '\n')
        
# hourly_activity('Madison Square Garden').iloc[0][0]
hourly_all_venues()
```
